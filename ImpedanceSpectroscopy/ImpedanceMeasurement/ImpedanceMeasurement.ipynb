# Impedance Measurement

The aim of impedance spectroscopy is to characterize the electrical properties of devices or materials by the impedance complex function $Z^*(\omega)$. It is defined by the ratio of the voltage between two electrical ports of a sample object and the resulting current through the ports.

According to this definition, impedance is similar to electric dc resistance which defines the ratio $\mathrm{U} / \mathrm{I}$ where $\mathrm{U}$ and $\mathrm{I}$ are constant in time. In an impedance measurement, however, an alternating ac voltage $\mathrm{U}_0$ with a fixed frequency $\omega / 2 \pi$ is applied to the sample under test. $\mathrm{U}_0$ causes a current $\mathrm{I}_0$ at the same frequency in the sample. In addition, there will generally be a phase shift between current and voltage described by the phase angle $\varphi$.

The ratio of $\mathrm{U}_0$ and $\mathrm{I}_0$ and the phase angle $\mathrm{j}$ are determined by the electrical properties of the sample. For simple calculation and representation of the formulas, it is convenient to use complex notation.
Voltage: $\quad U^*(t, \omega):=U_0 \cos (\omega t) \quad=\Re\left[U^*(\omega) \exp (j \omega t)\right]$
Current: $\quad I^*(t, \omega):=I_0 \cos (\omega t+\phi)=\mathfrak{R}\left[I^*(\omega) \exp (\omega t+\phi)\right]$
with

$$
U^*=U^{\prime}+j U^{\prime \prime}
$$

and

$$
I^*=I^{\prime}+j I^{\prime \prime} ; I_0=\sqrt{I^{\prime 2}+I^{\prime \prime 2}} ; \tan (\phi)=\frac{I^{\prime \prime}}{I^{\prime}}
$$

where $U^*, U^{\prime}, U^{\prime \prime}, U_0, I^*, I^{\prime}, I^{\prime \prime}, I_0$ and $\phi$ all are frequency dependent. For a sample with linear electrical response, the ratio of $\mathrm{U}^*(\omega)$ and $\mathrm{I}^*(\omega)$ does not depend on the magnitude $\mathrm{U}_0$ of the applied voltage and the measurement result can be reduced to one complex function which can be defined by several basic representations. The most common one is the impedance

$$
Z^*(\omega)=Z^{\prime}(\omega)+j Z^{\prime \prime}(\omega)=\frac{U^*(\omega)}{I^*(\omega)}
$$

Other often used representations are

- Admittance: $\quad Y^*(\omega):=\frac{1}{Z^*(\omega)}$
- Capacity: $\quad C^*(\omega):=\frac{1}{j \omega Z^*(\omega)}$
- Inductance: $\quad L^*(\omega):=\frac{Z^*(\omega)}{j \omega}$

These representations show the following properties.
For an ideal resistor or conductor with resistance $R$ and conductance $S=1 / R, Z^*(\omega)=R$, and $Y^*(\omega)=S$.
For an ideal capacitor with capacity $\mathrm{C}_0, \mathrm{C}^*(\omega)=\mathrm{C}_0$. If a material is placed between two electrodes, $\mathrm{C}^*(\omega)$ is directly proportional to the complex material permittivity $\varepsilon^*(\omega)$.

For an ideal inductor with inductance $\mathrm{L}_0, \mathrm{~L}^*(\omega)=\mathrm{L}_0$. If a material is placed into an inductive coil, $\mathrm{L}^*(\omega)$ is directly proportional to the complex material permeability $\mu^*(\omega)$.

The used representation depends on the sample type and the personal preference of the researcher. In practice, samples are never ideal. Nevertheless, one generally selects a representation close to the electrical sample type. E.g. a mainly capacitive sample generally is represented in terms of $C^*(\omega)$. Independent of the actual representation, one should keep in mind that they all contain the same information, i.e., the sample response current to an applied voltage.
Other types of frequently used representations are defined as ratios of real and imaginary parts of the complex data representations shown above. These are the loss factor

$$
\tan \delta:=-\frac{Z^{\prime}(\omega)}{Z^{\prime \prime}(\omega)}=-\frac{C^{\prime \prime}(\omega)}{C^{\prime}(\omega)}
$$

and the phase factor

$$
\tan \phi=\frac{1}{\tan \delta}
$$

$$
\begin{array}{|c|c|c|c|c|}
\hline \text { Case } & \boldsymbol{\varphi} & \boldsymbol{\operatorname { t a n }}(\boldsymbol{\varphi}) & \boldsymbol{\delta} & \boldsymbol{\operatorname { t a n } ( \boldsymbol { \delta } )} \\
\hline \text { Ideal resistor } & 0^{\circ} & 0 & 90^{\circ} & \infty \\
\hline \text { Ideal capacitor } & -90^{\circ} & -\infty & 0^{\circ} & 0 \\
\hline \text { Ideal inductor } & 90^{\circ} & \infty & 180^{\circ} & 0 \\
\hline
\end{array}
$$

In impedance spectroscopy, $\mathrm{Z}^*(\omega)$ is measured over a broad frequency range e.g. from $\mathrm{mHz}$ to several $\mathrm{GHz}$ in order to obtain as much information as possible.


Miniature radiofrequency (RF) coils have a variety of applications in magnetic resonance imaging (MRI) and spectroscopy (MRS). While they might appear useful only for chemical analysis of very small (nanolitre) volumes, they are increasingly used in human and animal studies. They have been used as fiducials to locate catheters and endoscopes, and also to obtain images from localized regions of tissue, aiming at micron-scale resolution. Another potential application is the spectroscopic evaluation of biopsy samples. Because devices are often single use, lowcost implementations are needed.
