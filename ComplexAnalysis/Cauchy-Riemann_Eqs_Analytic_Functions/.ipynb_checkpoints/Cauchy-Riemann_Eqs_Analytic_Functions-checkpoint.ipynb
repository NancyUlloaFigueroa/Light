# The Cauchy-Riemann Equations and Analytic Functions


Complex function theory really starts with the study of what it means to talk of the derivative of $\mathrm{f}(\mathrm{z})$. In real function theory, the derivative of $\mathrm{g}(\mathrm{x})$ at $\mathrm{x}=\mathrm{x}_0$ is defined as

$$
\left.\frac{\mathrm{dg}}{\mathrm{dx}}\right|_{\mathrm{x} \rightarrow \mathrm{x}_0}=\lim _{\Delta \mathrm{x} \rightarrow 0} \frac{\mathrm{g}\left(\mathrm{x}_0+\Delta \mathrm{x}\right)-\mathrm{g}\left(\mathrm{x}_0\right)}{\Delta \mathrm{x}}=\mathrm{g}^{\prime}\left(\mathrm{x}_0\right) .
$$

We do almost the same thing with a complex function. Indeed, the formal definition for the derivative of a complex $f(z)$ at $z=z_0$ is

$$
\left.\frac{\mathrm{df}}{\mathrm{dx}}\right|_{\mathrm{z}-\mathrm{z}_0}=\lim _{\Delta \mathrm{z} \rightarrow 0} \frac{\mathrm{f}\left(\mathrm{z}_0+\Delta \mathrm{z}\right)-\mathrm{f}\left(\mathrm{z}_0\right)}{\Delta \mathrm{z}}=\mathrm{f}^{\prime}\left(\mathrm{z}_0\right) .
$$

The vanishing of $\Delta z=\Delta x+i \Delta y$ is, however, not quite as straightforward as it is in the case of a real variable. In that simpler case, where we let $\Delta x \rightarrow 0$ to calculate $\mathrm{g}^{\prime}\left(\mathrm{x}_0\right), \Delta \mathrm{x}$ only has to vanish along the one-dimensional real axis. That is, $\Delta \mathrm{x}$ can shrink to zero in just two ways: either from the left of $x_0$ or from the right of $x_0$. But in the complex case we must take into account that, since $z_0$ is a point in the complex, two-dimensional plane, then $\Delta z$ can shrink to zero in an infinity of different ways (from the left of $\mathrm{z}_0$, from the right of $\mathrm{z}_0$, from below $\mathrm{z}_0$, from above $z_0$ or, indeed, from any direction of the compass). So, just how does $\Delta z \rightarrow 0$ ?
Mathematicians consider the most condition-free definition possible for the derivative to be the best definition, and so their answer to our question is: we want $f^{\prime}\left(z_0\right)$ to be the same independent of how $\Delta z \rightarrow 0$. To have this be the case, as you might suspect, comes with a price. If $f=u+i v$ then the price for a derivative at you might suspect, comes with a price. If $\mathrm{f}=\mathrm{u}+\mathrm{iv}$ then the price for a derivative at $\mathrm{z}=\mathrm{z}_0$ that doesn't depend on the precise nature of how $\Delta \mathrm{z} \rightarrow 0$ is that $\mathrm{u}$ and $\mathrm{v}$ $\mathrm{c}=\mathrm{z}_0$ that doesn't depend on the precise nature of how $\Delta \mathrm{z} \rightarrow 0$ is that $\mathrm{u}$ and $\mathrm{V}$ these conditions are satisfied at $\mathrm{z}=\mathrm{z}_0$ and at all points in a region (domain or neighborhood are terms that are also used) surrounding $\mathrm{z}_0$, then we say that $\mathrm{f}(\mathrm{z})$ is an analyric function in that region.

The conditions for $\mathrm{f}(\mathrm{z})$ to be analytic are called the Cauchy-Riemann (C-R) equations, which are actually pretty easy to state: at $\mathrm{z}=\mathrm{z}_0$ it must be true that

$$
\frac{\partial u}{\partial x}=\frac{\partial v}{\partial y}
$$

and

$$
\frac{\partial u}{\partial y}=-\frac{\partial v}{\partial x}
$$

# Analyticity

Whenever $f^{\prime}\left(z_0\right)$ exists, $f$ is said to be analytic (or regular, or holomorphic) at the point $z_0$. The function is analytic throughout a region in the complex plane if $f^{\prime}$ exists for every point in that region. Any point at which $f^{\prime}$ does not exist is called a singularity or singular point of the function $f$.
If $f(z)$ is analytic everywhere in the complex plane, it is called entire.

__Examples__
- $1 / z$ is analytic except at $z=0$, so the function is singular at that point.
- The functions $z^n, n$ a nonnegative integer, and $e^z$ are entire functions.

# The Cauchy-Riemann Conditions

The Cauchy-Riemann conditions are necessary and sufficient conditions for a function to be analytic at a point.

Suppose $f(z)$ is analytic at $z_0$. Then $f^{\prime}\left(z_0\right)$ may be obtained by taking $\delta z$ to zero through purely real, or through purely imaginary values, for example.

If $\delta z=\delta x, \delta x$ real, we have, upon writing $f$ in terms of its real and imaginary parts, $f=u+i v$,

$$
f^{\prime}\left(z_0\right)=\left(\frac{\partial u}{\partial x}+i \frac{\partial v}{\partial x}\right)_{z=z_0} .
$$

On the other hand, if $\delta z=i \delta y, \delta y$ real, we have similarly,

$$
\begin{aligned}
f^{\prime}\left(z_0\right) & =\left(\frac{\partial u}{i \partial y}+i \frac{\partial v}{i \partial y}\right)_{z=z_0} \\
& =\left(-i \frac{\partial u}{\partial y}+\frac{\partial v}{\partial y}\right)_{z=z_0} .
\end{aligned}
$$

Since the derivative is independent of how the limit is taken, we can equate these two expression, meaning that they must have equal real and imaginary parts,

$$
\frac{\partial u}{\partial x}=\frac{\partial v}{\partial y}, \quad \frac{\partial v}{\partial x}=-\frac{\partial u}{\partial y} .
$$

These are the Cauchy-Riemann conditions.

These conditions are not only necessary, but if the partial derivatives are continuous, they are sufficient to assure analyticity. Write

$$
\begin{aligned}
f(z+\delta z)-f(z)= & u(x+\delta x, y+\delta y)-u(x, y)+i[v(x+\delta x, y+\delta y)-v(x, y)] \\
= & u(x+\delta x, y+\delta y)-u(x, y+\delta y)+u(x, y+\delta y)-u(x, y) \\
& +i[v(x+\delta x, y+\delta y)-v(x, y+\delta y)+v(x, y+\delta y)-v(x, y)] \\
= & \delta x \frac{\partial u}{\partial x}+\delta y \frac{\partial u}{\partial y}+i\left[\delta x \frac{\partial v}{\partial x}+\delta y \frac{\partial v}{\partial y}\right]
\end{aligned}
$$

which becomes, if the Cauchy-Riemann conditions hold

$$
\begin{aligned}
f(z+\delta z)-f(z) & =\delta x \frac{\partial u}{\partial x}-\delta y \frac{\partial v}{\partial x}+i\left[\delta x \frac{\partial v}{\partial x}+\delta y \frac{\partial u}{\partial x}\right] \\
& =(\delta x+i \delta y)\left[\frac{\partial u}{\partial x}+i \frac{\partial v}{\partial x}\right]
\end{aligned}
$$

so since $\delta z=\delta x+i \delta y$, we see

$$
\frac{\delta f}{\delta z} \rightarrow \frac{\partial u}{\partial x}+i \frac{\partial v}{\partial x}
$$

independently of how $\delta z \rightarrow 0$, so

$$
f^{\prime}(z)=\frac{\partial u}{\partial x}+i \frac{\partial v}{\partial x}
$$

exists.

>### Example
>
>Consider the function $z^*$ of $z$; that is, if $z=x+i y, z^*=x-i y$. The CauchyRiemann conditions never hold,
>
>$$
\frac{\partial x}{\partial x}=1 \neq \frac{\partial(-y)}{\partial y}=-1,
$$
>
>so $z^*$ is nowhere an analytic function of $z$.

>
>Suppose that $z=x+i y$. The complex-valued function $f(z)=z^2$ is differentiable at any point $z$ in the complex plane.
>

>

\end{aligned}

# Conformal mappings

First, the Cauchy-Riemann equations may be written in complex form

$$
i \frac{\partial f}{\partial x}=\frac{\partial f}{\partial y} .
$$

In this form, the equations correspond structurally to the condition that the Jacobian matrix denoted by $\mathbf{J}$, whose $(i, j)$ th entry is $\mathbf{J}_{i j}=\frac{\partial f_i}{\partial x_j}$, or explicitly:

$$
\mathbf{J}=\left[\begin{array}{lll}
\frac{\partial \mathbf{f}}{\partial x_1} & \cdots & \frac{\partial \mathbf{f}}{\partial x_n}
\end{array}\right]=\left[\begin{array}{c}
\nabla^{\mathrm{T}} f_1 \\
\vdots \\
\nabla^{\mathrm{T}} f_m
\end{array}\right]=\left[\begin{array}{ccc}
\frac{\partial f_1}{\partial x_1} & \cdots & \frac{\partial f_1}{\partial x_n} \\
\vdots & \ddots & \vdots \\
\frac{\partial f_m}{\partial x_1} & \cdots & \frac{\partial f_m}{\partial x_n}
\end{array}\right]
$$

is of the form

$$
\left(\begin{array}{cc}
a & -b \\
b & a
\end{array}\right),
$$

where $a=\partial u / \partial x=\partial v / \partial y$ and $b=\partial v / \partial x=-\partial u / \partial y$. A matrix of this form is the matrix representation of a complex number. Geometrically, such a matrix is always the composition of a rotation with a scaling, and in particular preserves angles. The Jacobian of a function $f(z)$ takes infinitesimal line segments at the intersection of two curves in $z$ and rotates them to the corresponding segments in $f(z)$. Consequently, a function satisfying the Cauchy-Riemann equations, with a nonzero derivative, preserves the angle between curves in the plane. That is, the Cauchy-Riemann equations are the conditions for a function to be conformal.

Moreover, because the composition of a conformal transformation with another conformal transformation is also conformal, the composition of a solution of the Cauchy-Riemann equations with a conformal map must itself solve the Cauchy-Riemann equations. Thus the Cauchy-Riemann equations are conformally invariant.
