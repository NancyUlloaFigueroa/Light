# N-Dimensional Complex Numbers

An acquaintance with the established two dimensional complex number system naturally gives rise to the question as to whether this could be extended to a complex number system in a higher number of dimensions, creating complex numbers of the form

$$
\mathbf{a}+\mathbf{i} \mathbf{b}+\mathbf{j} \mathbf{c}+\mathbf{k} \mathbf{d} \ldots+\mathbf{n n}
$$

in $\mathbf{n}+\mathbf{1}$ dimensions.

The usual answer to this question indicates that the only mathematically consistent first step in this direction requires the use of quaternions, which are represented as a kind of 4-dimensional complex number system. This answer indicates that a straightforward intermediate step to 3-dimensional complex numbers is a mathematical impossibility.

This notion of mathematical impossibility is supported by the Hurwitz so-called [1,2,4,8 theorem](https://en.wikipedia.org/wiki/Hurwitz%27s_theorem_(composition_algebras)), which proved that numbers associated with a numbers algebra, or 'normed division algebra' cannot occur in any other than in $1,2,4$, or 8 dimensions.

It follows from this that a straightforward attempt to develop a 3- or higher-dimensional complex number system, of dimensionality other than those mentioned, cannot succeed.

Before considering the method I shall describe here for the creation of 3- and other-dimensional complex numbers in a manner that does not violate the Hurwitz limitation (which therefore requires other than a simple, single algebra), Let's shall review the difficulties involved in making such an attempt.

# PROBLEMS WITH THE CREATION OF 3-DIMENSIONAL COMPLEX NUMBERS:

We shall first consider the apparently insoluble problems associated with 3 -dimensional complex numbers, as a first step to higher-dimensional forms, being represented, for example, by a number of the form

$$
\mathbf{x}+\mathbf{i y}+\mathbf{j z}
$$

With regard to a 3-dimensional form like $\mathbf{x}+\mathbf{i} \mathbf{y}+\mathbf{j z}$, should we not expect to say that, just as the operator $\mathbf{i}$ rotates $\mathbf{y}$ from the $\mathbf{x}$ to the $\mathbf{i}$ direction, the operator $\mathbf{j}$ similarly rotates $\mathbf{z}$ from the $\mathbf{x}$ to the $\mathbf{j}$ direction, orthogonal to both the $\mathbf{x}$ and $\mathbf{i}$ directions? Here, the complex number has three independent, or orthogonal components, represented by three orthogonal directions in a 3-dimensional complex number space. This is the straightforward extrapolation from 2 to 3 dimensions, assuming that the operator $\mathbf{j}$ operates orthogonally to $\mathbf{i}$, but is in all other respects identical to $\mathbf{i}$. Then, just as $\mathbf{i}^* \mathbf{i}=\mathbf{- 1}$, we have $\mathbf{j}^* \mathbf{j}=\mathbf{- 1}$.

A problem immediately arises, however. In the $3^{-}$ dimensional complex number space, we have three independent planes of rotation of vectors: the $\mathbf{x}, \mathbf{i}$ plane, the $\mathbf{x}, \mathbf{j}$ plane, and the $\mathbf{i}, \mathbf{j}$ plane. What, we may then ask, is to be said of the relationship of the operators $\mathbf{i}$ and $\mathbf{j}$ in respect of rotations in the $\mathbf{i}, \mathbf{j}$ plane?

For rotations in this plane to be consistent with those in the other two planes, we should have something like:
$\mathbf{j i y}=\mathbf{j y}$, and $\mathbf{j j y}=\mathbf{j j i y}=\mathbf{- y}$, as describing a rotation of $\mathbf{y}$ from the $\mathbf{i}$ to the $\mathbf{j}$ direction, and then from the $\mathbf{j}$ direction to the negative real axis. This, however, is clearly not consistent with the behavioural description of the operators as simply $\sqrt{-1}$. How can we say $\mathbf{j i}=\mathbf{j}$ ? How does the operator $\mathbf{i}$ disappear, in a manner consistent with the operator descriptions?

We might expect $\mathbf{j i}=\mathbf{- 1}$, which would rotate $\mathbf{y}$ in the $\mathbf{- x}$ direction, instead of the $\mathbf{j}$ direction, which is unsatisfactory. Also, jjy is, according to the operator $\mathbf{j}$, is just $\mathbf{- y}$, whereas jjiy might also be $\mathbf{- y}$, instead of $\mathbf{i y}$, if $\mathbf{j i}=\mathbf{j}$, which does not make sense.

One way out might be to adopt the strange result of forbidding ji to be resolved into anything simpler while, at the same time, requiring the interpretation that $\mathbf{j i}$ is in the same direction as $\mathbf{j}$. But then we have the problem about the possible relationship between ji and ij. Also, problems will arise due to the fact that $\mathbf{j i j i}=+\mathbf{1}$. Another possible resolution is to specify that $\mathbf{j}$ operates only on the parameter of $\mathbf{i}$, and not on the operator $\mathbf{i}$ itself. Thus, no operations such as $\mathbf{j i}$ would be allowed to exist.

There is a further difficulty, however, in that the modulus of numbers must not be altered by operators themselves, which can be a problem when there are two operators $\mathbf{i}$, and $\mathbf{j}$. If an operation on a complex number $\mathbf{x}+\mathbf{i y}$ by the operator $\mathbf{j}$ changes the modulus, it creates an arithmetical effect, which is contrary to the definition of $\mathbf{j}$ as a non-arithmetical operator similar to the operator i. This would happen if you had

$$
\mathbf{j}(\mathrm{x}+\mathbf{i y})=\mathbf{j}(\mathrm{x}+\mathbf{y})
$$

Then

$$
\mathbf{j j}(\mathbf{x}+\mathbf{y})=-(\mathbf{x}+\mathbf{y})
$$

When $\mathbf{i}$ rotates $\mathbf{x}+\mathbf{i y}$ to $-\mathbf{y}+\mathbf{i x}$, the modulus

$$
|\mathbf{x}+\mathbf{i y}|=|-\mathbf{y}+\mathbf{i x}|
$$

is unchanged, and the same applies to $\mathbf{j}(\mathbf{x}+\mathbf{j z})$.

# The number i from the ordinary complex plane (Flat Land) does not exist in 3 dimensions!

On the impossibility of $X^2=-1$. A few times I mentioned that in the complex $\mathbb{R}^3$ space the number $i$ with $i^2=-1$ does not exist.

That has all kinds of far reaching consequences, if for example $i$ would exist in $\mathbb{R}^3$ space, in that case it would not be 3 dimensional.

So we are going to give proof to the fact that $\forall X \in \mathbb{R}^3: X^2 \neq-1$.
If we take as usual $X=(x, y, z)$ we get $X^2=\left(x^2-2 y z,-z^2+2 x y, y^2+2 x z\right)$ and if we try to solve $X^2=-1$ that is equivalent to:

$$
\left\{\begin{array}{rlr}
x^2-2 y z & =-1 \\
-z^2+2 x y & =0 \\
y^2+2 x z & =0
\end{array}\right.
$$

multiply the second and third row by $z$ and $y$ :

$$
\left\{\begin{aligned}
x^2-2 y z & =-1 \\
-z^3+2 x y z & =0 \\
y^3+2 x y z & =0
\end{aligned}\right.
$$

That is equal to

$$
\left\{\begin{aligned}
x^2-2 y z & =-1 \\
y^3 & =-z^3
\end{aligned}\right.
$$

Because $y^3=-z^3$ simply means $y=-z$ we can substitute $y=-z$ in the first row $x^2-2 y z=-1$, this becomes $x^2+2 z^2=-1$.

And that my dear reader is not possible with real numbers...

# Are quaternions more elegant than complex numbers for theory?

Absolutely, unequivocally, not. Complex numbers are one of the cornerstones of modern mathematics, whereas the Quaternions are a way to generate division algebras. Division algebras are good for some stuff, like certain formulations of Class Field Theory, but honestly -- they're so much less significant than the complex numbers that it's ridiculous. They have some "applications" in physics, but those can all be formulated without the quaternions. They're just helpful notation. The complex numbers, however, carry depth because of: a. the fundamental theorem of algebra, and b. the (AMAZING!) theory that arises when you extend the definition of "derivative" to the theory of complex-valued functions. None of those things are interesting over the quaternions.


# what is the relation between quaternions and imaginary numbers?

Let's start with the real line $\mathbb{R}$ constituted of real numbers $t$. This field is not algebraically closed from the absence of a real solution to the simple polynomial equation :
$$
X^2=-1
$$

A solution of (1) named $\mathbf{i}$ was invented and supposed belonging to the (imaginary) orthogonal to the real line at 0 . The 'numbers' $t+x \mathbf{i}$ with $t, x$ real constituting the field $\mathbb{C}$. This field is algebraically closed and nothing further needed to be added at this point.

But Hamilton wanted to get out of the plane and imagined another independent solution $\mathbf{j}$ of (1) belonging to the orthogonal to the real and imaginary lines at 0 .
Playing with these new 'extended numbers' $t+\boldsymbol{x} \mathbf{i}+\boldsymbol{y} \mathbf{j}$ filling the ${ }_3 \mathrm{D}$ space he got into trouble with multiplication since the product of two 'extended numbers' contained the products $\mathbf{i} \mathbf{j}$ and  $\mathbf{j} \mathbf{i}$.

Suppose that $\mathbf{i j}$ belongs to the $3 \mathrm{D}$ space : this means that $\mathbf{i} \mathbf{j}=\boldsymbol{a}+\boldsymbol{b} \mathbf{i}+\boldsymbol{c} \mathbf{j}$ but if we multiply this (at the left) by $\mathbf{i}$, replace $\mathbf{i} \mathbf{j}$ by the initial expression and simplify everything then we get $\mathbf{j}$ as $\mathbf{j}=\boldsymbol{\alpha}+\boldsymbol{\beta} \mathbf{i}:$ the $3 \mathrm{D}$ space collapsed back to the initial complex plane!

From the distributivity and associativity of the product we can't thus have two independent $\mathbf{i}$ and $\mathbf{j}$ without their product $\mathbf{k}:=\mathbf{i} \mathbf{j}$ being independent too : the $4 \mathrm{D}$ space generated by $(\mathbf{1}, \mathbf{i}, \mathbf{j}, \mathbf{k})$ is mandatory!

And what about $\mathbf{j} \mathbf{i}$ ?

If we suppose that $\mathbf{i j}=\mathbf{j} \mathbf{i}$ then everything will remain commutative with $\mathbf{k}^2=\mathbf{1}$ and we obtain what was named tessarines by James Cockle and later bicomplex numbers that were further generalized to multicomplex numbers ( $\mathbf{k}$ itself defining a split-complex number).

But before all this Hamilton noticed that $(x \mathbf{i}+y \mathbf{j})^2=x y(\mathbf{i} \mathbf{j}+\mathbf{j} \mathbf{i})-\left(x^2+y^2\right)$ so that abandoning commutativity and supposing that $\mathbf{k}=\mathbf{i j}=-\mathbf{j} \mathbf{i}$ he could rewrite not only $(x \mathbf{i}+y \mathbf{j})^2=-\left(x^2+y^2\right)$ but also $\mathbf{k}^2=-1$ and $(x \mathbf{i}+y \mathbf{j}+z \mathbf{k})^2=-\left(x^2+y^2+z^2\right)$.

The 3D space he wished was thus generated by $(\mathbf{i}, \mathbf{j}, \mathbf{k})$ instead of $(\mathbf{1}, \mathbf{i}, \mathbf{j})$ and he gained further an interesting fourth dimension! He was proud enough of his discovery to carve his very nice result into the stone of Brougham Bridge:

$$
\mathbf{i}^2=\mathbf{j}^2=\mathbf{k}^2=\mathbf{i j k}=-1
$$
