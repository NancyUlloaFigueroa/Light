## Exersice.
 
> Suppose that $A=(0,0), B=(1,1),$ and that $h(x, y)=x y$. Calculi the contour integral along the curve $\mathrm{C}_{1}$. Repeat the calculi along the curve $\mathrm{C}_{2}$
>

> <center> Fig. $\quad$ Two different line integral paths. </center>
> <br><br/>

## Answer.
> To start, let's suppose that $\mathrm{C}=\mathrm{C}_{1}$ is the broken path shown in Fig. The first part is along the $\mathrm{x}$-axis from $\mathrm{x}=0$ to $\mathrm{x}=1$, and then the second part is straight-up from $\mathrm{x}=1(\mathrm{y}=0)$ to $\mathrm{x}=1$ $(y=1)$. So, but for this path $y=0$ along the $x$-axis and so $h(x, y)=0)$, and $\mathrm{x}=1$ on the vertical portion of $\mathrm{C}_{1}$ [and so $\mathrm{h}(\mathrm{x}, \mathrm{y})=\mathrm{y}$]. Thus, the two line integrals on this path are:
>
>$$
\mathrm{I}_{\mathrm{x}}=\int_{\mathrm{C}_{1}} \mathrm{h}(\mathrm{x}, \mathrm{y}) \mathrm{d} \mathrm{x}=\int_{0}^{1} 0 \mathrm{d} \mathrm{x}+\int_{1}^{1} \mathrm{y} \mathrm{d} \mathrm{x}=0+0=0
$$
>  
> and
>  
>$$
\mathrm{I}_{\mathrm{y}}=\int_{\mathrm{C}_{1}} \mathrm{h}(\mathrm{x}, \mathrm{y}) \mathrm{d} \mathrm{y}=\int_{0}^{0} 0 \mathrm{d} \mathrm{y}+\int_{0}^{1} \mathrm{ydy}=\left.\left(\frac{1}{2} \mathrm{y}^{2}\right)\right|_{0} ^{1}=\frac{1}{2}
$$
> 

> Along the path $\mathrm{C}_{2},$ on the other hand, having $\mathrm{y}=\mathrm{x}$ from $\mathrm{A}$ to $\mathrm{B},$ and $\mathrm{so}$ $h(x, y)=x^{2}$ [or, equivalently,  $y^{2}$]. So, on this path the line integrals are
>  
>$$
\mathrm{I}_{\mathrm{x}}=\int_{\mathrm{C}_{2}} \mathrm{h}(\mathrm{x}, \mathrm{y}) \mathrm{d} \mathrm{x}=\int_{0}^{1} \mathrm{x}^{2} \mathrm{dx}=\left.\left(\frac{1}{3} \mathrm{x}^{3}\right)\right|_{0} ^{1}=\frac{1}{3}
$$
>  
> and
>  
>$$
\mathrm{I}_{\mathrm{y}}=\int_{\mathrm{C}_{2}} \mathrm{h}(\mathrm{x}, \mathrm{y}) \mathrm{d} \mathrm{y}=\int_{0}^{1} \mathrm{y}^{2} \mathrm{d} \mathrm{y}=\left.\left(\frac{1}{3} \mathrm{y}^{3}\right)\right|_{0} ^{1}=\frac{1}{3}
$$
> 
> Clearly, the values of the $I_{x}, I_{y}$ line integrals are path-dependent and, for a given path, the $I_{x}, I_{y}$ line integrals may or may not be equal. It can combined the $I_{x}$ and $I_{y}$ line integrals to write the line integral along $\mathrm{C}$ as $\mathrm{I}_{\mathrm{C}}=\mathrm{I}_{\mathrm{x}}+i \mathrm{I}_{\mathrm{y}},$ and so $\mathrm{I}_{\mathrm{C}_{1}}=i_{2}^{\mathrm{l}}$ while
> 
>$$
\mathrm{I}_{\mathrm{C}_{2}}=\frac{1}{3}+i \frac{1}{3}
$$
> 

> Notice that in Fig. it could be written the unbroken line segment $\mathrm{AB}$ as $\mathrm{z}=\mathrm{x}+$ iy or, as $\mathrm{y}=\mathrm{x}, \mathrm{z}=\mathrm{x}+i \mathrm{x}=\mathrm{x}(1+i)$ and so
$\mathrm{dz}=(1+i) \mathrm{dx}$. Then, as $\mathrm{h}(\mathrm{x}, \mathrm{y})=\mathrm{h}(\mathrm{x}, \mathrm{x})=\mathrm{x}^{2},$ it gets
>
>$$
\mathrm{I}_{\mathrm{C}_{2}}=\int_{0}^{1} \mathrm{x}^{2}(1+i) \mathrm{d} \mathrm{x}=\left.(1+i)\left(\frac{1}{3} \mathrm{x}^{3}\right)\right|_{0} ^{1}=\frac{1}{3}+i \frac{1}{3}
$$
>  
> which is just as it has been calculated before.

### Kepler's Second law with Complex Variable

Imagine a point mass $m$ moving in a plane along the path given by [Eq. X37],

$$
z(t)=r(t) e^{i \theta(t)}
$$ [Eq. X37]

where now $z, r$ and $\theta$ are specifically indicated to be functions of time (t). The motion of $\mathrm{m}$ is due entirely to a force acting along the line connecting the mass to the source of the force: the classic example of this situation is the Earth (the 'point' mass) moving under the influence of the gravitational field of the Sun (which it'll be taken as being at the origin of the $x-y$ coordinate system). The attractive force on the Earth is, of course, always directed radially inward towards the Sun.

If the magnitude of the force are written on $\mathrm{m}$ as $\mathrm{f}(\mathrm{r}, \theta)$, Newton's famous second law of motion ('force is mass times acceleration') says

$$
\mathrm{f}(\mathrm{r}, \theta) \mathrm{e}^{i \theta}=\mathrm{m} \frac{\mathrm{d}^2 \mathrm{z}}{\mathrm{dt}^2} .
$$ [Eq. X38]

From [Eq. X37] it has

$$
\frac{\mathrm{dz}}{\mathrm{dt}}=\frac{\mathrm{dr}}{\mathrm{dt}} \mathrm{e}^{i \theta}+i \mathrm{r} \frac{\mathrm{d} \theta}{\mathrm{dt}} \mathrm{e}^{i \theta}
$$

and so

$$
\frac{\mathrm{d}^2 \mathrm{z}}{\mathrm{dt}^2}=\frac{\mathrm{d}^2 \mathrm{r}}{\mathrm{dt}^2} \mathrm{e}^{i \theta}+\frac{\mathrm{dr}}{\mathrm{dt}} i \frac{\mathrm{d} \theta}{\mathrm{dt}} \mathrm{e}^{i \theta}+i\left[\frac{\mathrm{dr}}{\mathrm{dt}} \frac{\mathrm{d} \theta}{\mathrm{dt}} \mathrm{e}^{i \theta}+\mathrm{r} \frac{\mathrm{d}^2 \theta}{\mathrm{dt}^2} \mathrm{e}^{i \theta}+\mathrm{r} \frac{\mathrm{d} \theta}{\mathrm{dt}} i \frac{\mathrm{d} \theta}{\mathrm{dt}} \mathrm{e}^{i \theta}\right]
$$

or,

$$
\frac{\mathrm{d}^2 \mathrm{z}}{\mathrm{dt}^2}=\left[\frac{\mathrm{d}^2 \mathrm{r}}{\mathrm{dt}^2}-\mathrm{r}\left(\frac{\mathrm{d} \theta}{\mathrm{dt}}\right)^2\right] \mathrm{e}^{i \theta}+i\left[2 \frac{\mathrm{dr}}{\mathrm{dt}} \frac{\mathrm{d} \theta}{\mathrm{dt}}+\mathrm{r} \frac{\mathrm{d}^2 \theta}{\mathrm{dt}^2}\right] \mathrm{e}^{i \theta} .
$$ [Eq. X39]

Using [Eq. X39] in [Eq. X38] and cancelling all the $\mathrm{e}^{i t}$ (which are never zero), it arrives at 

$$
\mathrm{f}(\mathrm{r}, \theta)=\mathrm{m}\left[\frac{\mathrm{d}^2 \mathrm{r}}{\mathrm{dt}^2}-\mathrm{r}\left(\frac{\mathrm{d} \theta}{\mathrm{dt}}\right)^2\right]+i \mathrm{~m}\left[2 \frac{\mathrm{dr}}{\mathrm{dt}} \frac{\mathrm{d} \theta}{\mathrm{dt}}+\mathrm{r} \frac{\mathrm{d}^2 \theta}{\mathrm{dt}^2}\right] .
$$

Equating real and imaginary parts of this last expression gives the famous differential equations of motion in a radial force field:

$$
f(r, \theta)=m\left[\frac{d^2 r}{d t^2}-r\left(\frac{d \theta}{d t}\right)^2\right]
$$ [Eq. X40]

and

$$
2 \frac{\mathrm{dr}}{\mathrm{dt}} \frac{\mathrm{d} \theta}{\mathrm{dt}}+\mathrm{r} \frac{\mathrm{d}^2 \theta}{\mathrm{dt}^2}=0 .
$$ [Eq. X41]

Interestingly, the result in [Eq. 41] was implicitly known long before Newton. Mathematics alone tells us that

$$
\frac{\mathrm{d}}{\mathrm{dt}}\left\{\mathrm{r}^2 \frac{\mathrm{d} \theta}{\mathrm{dt}}\right\}=2 \mathrm{r}\left(\frac{\mathrm{dr}}{\mathrm{dt}}\right) \frac{\mathrm{d} \theta}{\mathrm{dt}}+\mathrm{r}^2 \frac{\mathrm{d}^2 \theta}{\mathrm{dt}^2}=\mathrm{r}\left[2\left(\frac{\mathrm{dr}}{\mathrm{dt}}\right) \frac{\mathrm{d} \theta}{\mathrm{dt}}+\mathrm{r} \frac{\mathrm{d}^2 \theta}{\mathrm{dt}^2}\right]
$$

and, since the expression in the square brackets is zero by [Eq. X41] , it has

$$
\frac{d}{d t}\left\{r^2 \frac{d \theta}{d t}\right\}=0 .
$$

Thus, integration that gives

$$
\mathrm{r}^2 \frac{d \theta}{\mathrm{dt}}=\mathrm{C}
$$ [Eq. X42]

where $\mathrm{C}$ is a constant. This result has a historically important physical interpretation in the theory of planetary motion.

Look at Fig. X31, which shows a planet's location at times $t$ and $t+\Delta t$, with the Sun at the origin of the coordinate system. it is assumed $\Delta t$ is so small that the angular change $\Delta \theta$ in the radius vector's angle is also very small, and that the length of the radius vector remains essentially unchanged. Then, the area between the two dashed lines is essentially that of an isosceles triangle with area $\Delta \mathrm{A}$ given by

$$
\Delta \mathrm{A}=\frac{1}{2} \text { base times height } \approx \frac{1}{2}(\mathrm{r} \Delta \theta) \mathrm{r}=\frac{1}{2} \mathrm{r}^2 \Delta \theta .
$$

Dividing through by $\Delta \mathrm{t}$ it gives

<br><br/>
<p align="center"><img src="../Countur_Integrals___[images]/interpreting_planetary_motion.png" style="width:50%"></p>
<center> Fig. X31$\quad$ Interpreting $r^2 \frac{d \theta}{d t}$. </center>
<br><br/>

$$
\frac{\Delta A}{\Delta t} \approx \frac{1}{2} r^2 \frac{\Delta \theta}{\Delta t},
$$

an expression that becomes exact as $\Delta t \rightarrow 0$. That is, replacing the delta quantities with differential ones, it has

$$
\frac{\mathrm{dA}}{\mathrm{dt}}=\frac{1}{2} \mathrm{r}^2 \frac{\mathrm{d} \theta}{\mathrm{dt}}
$$

or, from [Eq. X42],

$$
\mathrm{r}^2 \frac{\mathrm{d} \theta}{\mathrm{dt}}=2 \frac{\mathrm{dA}}{\mathrm{dt}}=\mathrm{C} .
$$

This last expression is the mathematical form of the statement (given in 1609) by the German astronomer Johann Kepler (1671-1630) of his famous area law: the line joining the Sun to a planet sweeps over equal areas in equal time intervals. Kepler deduced this (the second of three general laws he discovered) not by physics or complex function theory, but rather from years of tedious observational data.
