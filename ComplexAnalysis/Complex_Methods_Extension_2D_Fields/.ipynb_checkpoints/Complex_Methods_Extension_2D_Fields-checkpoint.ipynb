# Some Extensions of Complex Methods for 2D Fields
### [Atomic Energy]



# Introduction

The advent of practical superconducting magnets has given renewed interest to general methods of analysis for two-dimensional fields, especially in connection with beam handing and accelerator applications.

Complex variable methods have long been used to set up potential functions that satisfy Laplace's equation in two dimensions. Maxwell already devoted a chapter to this subject. The transformations developed independently by Schwarz and Christoffel add a powerful tool.

The complex methods sumarized in this notes are based directly on the Cauchy-Riemann equations rather than on the Laplace's equations that can be derived from them. This approach tends to deal primarily with flelds rather than potentials. It permits us, for example, to set up a general integral formula for the field inside as well as outside a conductor with uniform current density. More generally, it helps to elucidate the far-reaching structural identity between analytic functions and static two-dimensional magnetic or electric fields.

Currents and the forces on them appear as residues of analytic functions.

A formula can be given for the discontinuity in complex field across an arbitrary cylindrical current sheet. Based on this "current sheet theorem" it has been possible, for example, to specify the current distribution on a circular or elliptic cylinder to produce any required field within the cylinder and even to specify a second outer cylinder which will simultaneously cancel all external field.

Any two-dimensional field configuration can be transformed into a plane whose coordinates are vector and scalar potentials and in which areas are proportional to field energy.

Complex methods such as those under discussion obtain added practical value from the fact that large computers can deal with complex variables directly so that prior separation into real and imaginary parts is not necessary.

# Analytic Functions and Complex Fields

For any region of the complex $Z=x+i Y$ plane the statement

   -(a) $F(Z)=U+i V$ is an analytic function of $Z$

is more powerful than either or both Laplace equations:

   -(b) $\nabla^2 \mathrm{U}=0$ and $\nabla^2 \mathrm{~V}=0$

since (b) follows from (a) but (a) does not follow from (b). If (a) could be deduced from (b) then both $U+i V$ and $U-i V$ would be analytic functions of $\mathrm{Z}$, which is in general not possible. In fact (a) includes information about the interrelation between the functions $U(X, Y)$ and $V(X, Y)$ which is not retained in (b).

These interrelations are specified by the two Cauchy-Riemann equations

$$
\frac{\partial U}{\partial X}=\frac{\partial V}{\partial Y} \quad \text { and } \quad \frac{\partial U}{\partial Y}=-\frac{\partial V}{\partial X}
$$ [Eq. 1]

which are equivalent to, i.e., necessary and sufficient for, statement (a).

A two-dimensional magnetostatic field is governed by the two Maxwell equations

$$
\frac{\partial B_X}{\partial X}+\frac{\partial B_Y}{\partial Y}=0 \quad \text { and } \quad \frac{\partial H_Y}{\partial X}-\frac{\partial H_Y}{\partial Y}=4 \pi \sigma
$$ [Eq. 2]

When the permeability, $\downarrow=\mathrm{B} / \mathrm{H}$, is constant and the current density, $\sigma$, is zero, the Maxwell relations [Eq. 2] are exactly the Cauchy-Riemann conditions [Eq. 1] for the complex combination of field components

$$
H(X, Y) \equiv H_Y+i H_X=B(X, Y) / \mu
$$ [Eq. 3]

to be an analytic function $F(Z)$ of $Z$ .

More generally, when the current density, $\sigma$, is not zero but is constant we see, by setting $U=H_Y-2 \pi \sigma X$ and $V=H_X+2 \pi \sigma Y$, that Maxwell's [Eq 2] constitute the Cauchy-Riemann equations for the statement

$$
F(Z)=H(X, Y)-2 \pi \sigma Z^*=\text { analytic function of } Z
$$ [Eq. 4]

where $Z^* = X - i Y$. Obviously, this more general statement includes the former special case for $\sigma=0$.

In a similar way it can be seen that the two Maxwell equations governing electrostatic fields

$$
\frac{\partial D_X}{\partial X}+\frac{\partial D_Y}{\partial Y}=4 \pi \rho \quad \text { and } \quad \frac{\partial E_Y}{\partial X}-\frac{\partial E_X}{\partial Y}=0
$$ [Eq. 5]

when $\epsilon=D / E$ and $\rho=$ charge density are real constants, constitute the Cauchy-Riemann conditions for

$$
\mathrm{F}(\mathrm{Z})=\mathrm{D}(\mathrm{X}, \mathrm{Y})+2 \pi i \rho \mathrm{z}^*=\text { analytic function of } \mathrm{Z}
$$ [Eq. 6]

where, analogous to the definition  [Eq. 3], we define

$$
\mathrm{D}(\mathrm{X}, \mathrm{Y})=\mathrm{D}_{\mathrm{Y}} + i \mathrm{D}_{\mathrm{X}} = \varepsilon  \mathrm{E}(\mathrm{X}, \mathrm{Y}) .
$$ [Eq. 7]

# Integral Formula for $F(Z)$

The Figure shows a normal cross section of a cylindrical conductor. Let the current density, $\sigma$, be constant within the boundary $\mathrm{C}$ and zero outside. Denote the points of the boundary by small $z$ and fleld points by capital $z$. Then the integral, taken in the positive sense around the boundary,

$$
F(z)=i \sigma \oint_C \frac{z^* d z}{z-z}
$$ [Eq. 8]

yields separate analytic functions for the $Z$ regions inside and outside the boundary such that the respective

complex fields, $\mathrm{H} \equiv \mathrm{H}_{\mathrm{Y}}+\mathrm{iH}_{\mathrm{X}}$, as defined in (3), are given by:

$$
\begin{array}{ll}
H_{\text {out }}=F(Z) & \text { for } Z=z_{\text {out }} \\
H_{\text {in }}=F(Z)+2 \pi \sigma Z^* & \text { for } Z=z_{\text {in }}
\end{array}
$$ [Eq. 9]

It can be shown that both expressions converge to the same field values at the surface of the conductor.

The corresponding electric fields due to uniform charge density $\rho$ within a cylinder are obtained by replacing i $\sigma$ by $\rho$.

# Current Filaments

The complex field due to a current filament $I$ at $z=x+$ iy is easily seen to be

$$
\mathrm{H}=\mathrm{H}_{\mathrm{Y}}+1 \mathrm{H}_{\mathrm{X}}=\frac{2 \mathrm{I}}{\mathrm{Z}-\mathrm{z}}
$$ [Eq. 11]

since, with $r=|z-z|$, the real and imaginary parts of (1I) give the usual expressions

$$
\mathrm{H}_{\mathrm{Y}}=\frac{2 \mathrm{I}}{\mathrm{r}^2}(\mathrm{X}-\mathrm{x}) \text { and } \mathrm{H}_{\mathrm{X}}=-\frac{2 \mathrm{I}}{\mathrm{r}^2}(\mathrm{Y}-\mathrm{y}) \text {. }
$$

Thus, in any field, an isolated current filament I constitutes a simple pole with residue $2 I$. The contour integral enclosing any number of filaments $I_n$ is $2 \pi i$ times the sum of the residues, that is $4 \pi i$ times the total current enclosed:

$$
\oint \mathrm{H}(\mathrm{Z}) \mathrm{dZ}=4 \pi \mathrm{i} \mathrm{I}_{\mathrm{n}} .
$$ [Eq. 12]

A distribution of current density, $\sigma(x, y)$, within a cylindrical conductor of arbitrary shape may be regarded as a bundle of infinitesimal current filaments. Therefore, on the basis of (11), the field outside the conductor is

$$
\mathrm{H}_{\text {out }}(z)=2 \iint \frac{\sigma(\mathrm{x}, \mathrm{y})}{Z-z} \mathrm{dxdy}
$$ [Eq. 13]

where the integration covers the conductor cross section.

For uniform current density, $\sigma=$ constant, closed expressions for $H_{\text {out }}(2)$ have been evaluated for polygonal, rectangular, and ribbon conductors. It can be shown that for constant $\sigma$ (13) is equivalent to $\mathrm{H}_{\text {out }}(\mathrm{Z})$ as given by (8) and (9).

# Forces on Currents

If a current filament $I$ is located at $z$ in a field $h(z)=h_Y+i h_X$ due to other sources, then the total field, by superposition, is

$$
H(Z)=h(Z)+\frac{2 I}{Z-z}
$$ [Eq. 14]

where $h(Z)$ is regular at $z$. The residue of $H^2$ at $z$ is easily seen to be $4 \mathrm{Ih}(z)$.

The components of force acting on unit length of the filament I at $z$ are

$$
\mathrm{E}_{\mathrm{X}}=-\mathrm{I \mu h_{ \textrm {Y } }} \text { and } \mathrm{f}_{\mathrm{Y}}=\mathrm{I \mu h}_{\mathrm{X}} \text {. }
$$ [Eq. 15]

We define the complex force per unit length as

$$
\mathrm{f} \equiv \mathrm{f}_{\mathrm{Y}}+\mathrm{if} \mathrm{f}_{\mathrm{X}}
$$ [Eq. 16]

so that

$$
\mathrm{f}=\mathrm{I} \mu\left(\mathrm{h}_{\mathrm{X}}-i \mathrm{~h}_{\mathrm{Y}}\right)=-\mathrm{i} \mu \mathrm{Ih}(\mathrm{z})
$$

which is $\mu / 4 i$ times the residue of $\mathrm{H}^2$ at $z$. When the permeability, $\mu$, is constant we can therefore find the resultant force per unit length on all the currents within a closed curve $\mathrm{C}$ as a contour integral ${ }^{8,9}$ :

$$
F=-\frac{\mu}{8 \pi} \oint_C H^2(Z) d Z=-\frac{1}{8 \pi} \oint_C B H d z .
$$ [Eq. 17]

The elementary calculation of $f$ as in (15) involves both $h(z)$ and $I$ but not the total field $\mathrm{H}(\mathrm{z})$ which has an infinite singularity at $\mathrm{z}$. Characteristically, the contour integral (17), by making use of the analytic properties of the field, involves only values of the total field $H(Z)$ on the contour $C$ and does not require separate knowledge of either h or $I$.

# Current Sheets

Consider a cylinder whose elements are perpendicular to the complex $Z$ plane at the points $z$ of curve $C$ in Fig. Let current dI flow upward along the elements of the cylinder that lie in the interval $\mathrm{dz}$. Then the analytic complex fields on the two sides of the curve, namely,

$$
\begin{aligned}
H(Z) & =H_R(Z) \text { to the right of } C \\
& =H_L(Z) \text { to the left of } C,
\end{aligned}
$$

approach limiting values on the curve $C$ whose difference is given by the following "current sheet theorem":

<br><br/>
<p align="center"><img src="Complex_Methods_Extension_2D_Fields___[images]/Cylindrical_Current_Sheet.png" style="width:30%"></p>
<center> Fig.$\quad$ Cylindrical current sheet. </center>
<br><br/>

$$
\mathrm{H}_{\mathrm{R}}(\mathrm{z})-\mathrm{H}_{\mathrm{L}}(\mathrm{z})=4 \pi \mathrm{i} \frac{\mathrm{dI}}{\mathrm{dz}}
$$ [Eq. 18]

Here dI is real, and right and left are defined by the direction chosen along $C$ for the complex increment $\mathrm{dz}$.
For example, the current distribution

$$
\frac{d I}{d \theta}=I_n \cos n \theta, \quad n=\text { positive integer }
$$ [Eq. 19]

along the elements of the right circular cylinder $z=r e^{i \theta}, x=$ constant, gives us

$$
\frac{d I}{d z}=\frac{I_n \cos n \theta}{i z}=\frac{I_n}{2 i z}\left[\left(\frac{z}{r}\right)^n+\left(\frac{r}{z}\right)^n\right]
$$

When there are no other currents, the external field, $\mathrm{H}_R(Z)$, must go to zero with $\mathrm{Z} \rightarrow \infty$ and the internal field, $\mathrm{H}_{\mathrm{L}}(\mathrm{Z})$, must remain finite at $\mathrm{z}=0$. Hence, using (18), we conclude that the fields are given by

$$
\mathrm{H}_{\mathrm{R}}(\mathrm{z})=\frac{2 \pi \mathrm{I}_n \mathrm{r}^{\mathrm{n}}}{\mathrm{z}^{\mathrm{n}+1}} \text { and } \mathrm{H}_{\mathrm{L}}(\mathrm{z})=-\frac{2 \pi \mathrm{I}_n \mathrm{Z}^{\mathrm{n}-1}}{\mathrm{r}^{\mathrm{n}}}
$$ [Eq. 20]

It will be seen that an arbitrary internal field,

$$
\mathrm{H}_{\mathrm{L}}(\mathrm{z})=\sum_{\substack{\infty \\ n=1}}^{\infty} \mathrm{H}_{\mathrm{n}} \mathrm{z}^{\mathrm{n}-1},
$$

where the $\mathrm{H}_n$ are complex constants, can be produced by appropriate Fourier superposition of $\cos n \theta$ and $\sin n \theta$ current distributions on a circular cylinder.

By using the same method we can find the required current distribution on any elliptic cylinder to produce any prescribed internal field. A second outer current sheet can be specified which will simultaneously cancel all outer fields. (A superconducting bending magnet with elliptical aperture is being designed).

# Current Sheet Forces

The resultant complex force per unit height, $f$, acting on an interval of current sheet from $z_1$ to $z_2$ can be evaluated from (17) by imagining the enclosing contour to be deformed from both sides toward coincidence with the current sheet:

$$
f=-\frac{\mu}{8 \pi} \int_{z_1}^{z_2}\left[H_R^2(z)-H_L^2(z)\right] d z, \mu=\text { constant } .
$$ [Eq. 21]

If we write

$$
\bar{B}(z)=\frac{\mu}{2}\left[\mathrm{H}_{\mathrm{R}}(\mathrm{z})+\mathrm{H}_{\mathrm{L}}(\mathrm{z})\right]
$$  [Eq. 22]

for the arithmetic mean induction at point $z$ of the sheet and multiply by (18), then (21 becomes

$$
f=-i \int_1^2 \bar{B} d I
$$

as it should.

In differential form (21) can also be written as

$$
\frac{\mathrm{d} f}{\mathrm{~d} z}=\frac{\mu}{8 \pi}\left[\mathrm{H}_{\mathrm{L}}^2(\mathrm{z})-\mathrm{H}_{\mathrm{R}}^2(\mathrm{z})\right]
$$  [Eq. 23]

Equations (21) and (23) enable us to calculate current sheet forces for an interval or at a point when the fields are known.

# Potentials and Field Energy

A two-dimensional magnetic field parallel to the $X, Y$ plane has vector and scalar potentials which are real functions of the coordinates:

$$
A=A(X, Y) \quad \text { and } \quad \Omega=\Omega(X, Y)
$$  [Eq. 24]

such that

$$
\mathrm{B}_{\mathrm{X}}=\frac{\partial \mathrm{A}}{\partial \mathrm{Y}}, \quad \mathrm{B}_{\mathrm{Y}}=-\frac{\partial \mathrm{A}}{\partial \mathrm{X}}, \quad \mathrm{H}_{\mathrm{X}}=-\frac{\partial \Omega}{\partial \mathrm{X}}, \quad \mathrm{H}_{\mathrm{Y}}=-\frac{\partial \Omega}{\partial \mathrm{Y}}
$$  [Eq. 25]

The scalar potential $\Omega$ exists in any simply connected region without currents; A exists everywhere.

Equations (24) may be regarded as a transformation from the X,Y plane to an $A, \Omega$ plane. The Jacobian of the transformation is

$$
J=\left|\begin{array}{ll}
\frac{\partial A}{\partial X} & \frac{\partial A}{\partial Y} \\
\frac{\partial \Omega}{\partial X} & \frac{\partial \Omega}{\partial Y}
\end{array}\right|=B_X H_X+B_Y H_Y=\vec{B} \cdot \vec{H}
$$  [Eq. 26]

which is in general not zero unless the field is zero. Therefore any simply connected $X, Y$ region $R$ without currents will be uniquely transformed into a region $R^{\prime}$ of the $A, \Omega$ plane whose area is

$$
S=\iint_{R^{\prime}} \mathrm{dA} \mathrm{d} \Omega=\iint_R \mathrm{~J} d \mathrm{X} \mathrm{dY}
$$  [Eq. 27]

where we have used the Jacobian to transform area integrals.
If the permeability, $\mu=\mathrm{B} / \mathrm{H}$, is a constant independent of $\mathrm{H}$ throughout $R$, then we see, from (26), that $J / 8 \pi$ is the energy density and, from (27), that $E_R=S / 8 \pi$ is the total energy in $R$ per unit thickness (normal to $X, Y$ ) of the field. Insofar as the $X, Y$ configuration can be subdivided into constant permeability regions the total energy is proportional to the total area of the $A, \Omega$ regions into which they are mapped by the transformation (24). Since the total field energy is always finite, the infinite $X, Y$ plane will be transformed into finite regions of the $A, \Omega$ plane. Field energies can be calculated as $A, \Omega$ areas.

The arguments of this section concerning field energy have not involved complex methods. However, it should be noted that, when $\mu$ is constant, the complex potential

$$
W(Z) \equiv-[(A / H)+i \Omega]=\int_{Z_0}^Z H d Z
$$  [Eq. 28]

is an analytic function of $\mathrm{z}$ in any simply connected region without current such that

$$
H=\frac{\mathrm{dW}}{\mathrm{d} Z}
$$  [Eq. 29]

is equivalent to the relations (25).

The fact that both potentials are required to form the analytic function (28) first led to the consideration of the transformation (24); for, when $i=1$, the $A, \Omega$ plane is the $-W$ plane, (24) is essentially the complex $W=W(Z)$ transformation, and $W$-plane areas are proportional to field energy.
